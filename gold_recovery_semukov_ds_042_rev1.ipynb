## Технологический процесс

Как золото получают из руды? Изучите внимательно стадии процесса.
Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

Опишем каждую стадию:

**1. Флотация**
- Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

**2. Очистка**

- Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

## Описание данных

Технологический процесс

- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- Xanthate **— ксантогенат (промотер, или активатор флотации);
- Sulphate — сульфат (на данном производстве сульфид натрия);
- Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

Параметры этапов

- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

Наименование признаков

Наименование признаков должно быть такое:

[этап].[тип_параметра].[название_параметра]

Пример: rougher.input.feed_ag

Возможные значения для блока [этап]:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

Возможные значения для блока [тип_параметра]:

- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### 1.2 Проверка расчётов эффективности обогащения

Проверим, что эффективность обогащения (процесс восстановления золота из золотосодержащей руды) чернового концентрата *rougher.output.recovery* рассчитана правильно. 

Вычислим ее на обучающей выборке gold_train для признака *rougher.output.recovery*. Найдем MAE между нашими расчетами *(переменная rougher.output.recovery.calc)* и значениями признака *rougher.output.recovery*. Опишем выводы.

Эффективность обогащения(восстановление золота из золотосодержащей руды) рассчитывается по формуле

**recovery = ((c*(f-t)) / (f*(c-t)))* 100**

- C — доля золота в концентрате после флотации - **rougher.output.concentrate_au**
- F — доля золота в сырье до флотации - **rougher.input.feed_au**
- T — доля золота в отвальных хвостах после флотации - **rougher.output.tail_au**

Для подсчета MAE между нашим результатом и значениями в датафрейме временно удалим пропуски в столбцах, используемых в расчете.

```python
#оценим количество значений Nan, которые предстоит исключить, для расчета формулы эффективности обогащения
print(gold_train['rougher.output.concentrate_au'].isna().sum())
print(gold_train['rougher.input.feed_au'].isna().sum())
print(gold_train['rougher.output.tail_au'].isna().sum())
```

Для подсчета MAE между нашим результатом и значениями в датафрейме временно удалим пропуски в столбцах, используемых в расчете.

```python
import warnings
warnings.filterwarnings('ignore')

# сохраним во временной таблице интересующие для расчета формулы данные и удалим пропуски методом dropna()
gold_train_dropna = gold_train.dropna(subset=['rougher.output.concentrate_au', 
                                              'rougher.input.feed_au',
                                              'rougher.output.tail_au', 
                                              'rougher.output.recovery'])

c = gold_train_dropna['rougher.output.concentrate_au']
f = gold_train_dropna['rougher.input.feed_au']
t = gold_train_dropna['rougher.output.tail_au']

#Расчитаем эффективность с помощью формулы recovery данной в инструкции.
def recovery(c,f,t):
    recovery = ((c*(f-t)) / (f*(c-t))) * 100
    return recovery

gold_train_dropna['rougher.output.recovery.calc'] = recovery(c,f,t)
    
mae_rougher_recovery = mean_absolute_error(gold_train_dropna['rougher.output.recovery'],
                                           gold_train_dropna['rougher.output.recovery.calc'])
print('MAE = ', mae_rougher_recovery)
```

Значение MAE стремится к нулю, следовательно, показатель рассчитан верно. Предположим, что и остальные показатели recovery рассчитаны верно. Попробуем заполнить признак rougher.output.recovery в обучающей выборке по имеющимся данным

### 3.1 Подготовка итоговой метрики  sMAPE

Напишем функцию для вычисления итоговой метрики sMAPE, где
- target    - реальные значения целевого признака
- predict - предсказания модели
- smape - симметричное среднее абсолютное процентное отклонение

```python
def smape(target, predict):
    return 1 / len(target) * sum( abs(target - predict) / ( (abs(target) + abs(predict)) * 0.5) ) * 100
```

Напишем функцию для вычисления итоговой метрики sMAPE, где
- final_smape - итоговое симметричное среднее абсолютное процентное отклонение


- rougher_target    - реальные значения целевого признака (эффективность обогащения после флотации)
- rougher_predict - предсказания модели (финальная эффективность обогащения)


- final_target    - реальные значения целевого признака (эффективность обогащения после флотации)
- final_predict - предсказания модели (финальная эффективность обогащения)

```python
def final_smape(smape_rougher, smape_final):

    return 0.25 * smape_rougher + 0.75 * smape_final
```

Нам необходимо спрогнозировать две величины: 
- эффективность обогащения чернового концентрата **rougher.output.recovery**
- эффективность обогащения финального концентрата **final.output.recovery**

- Модели которые будем использовать:  LinearRegression, RandomForestRegressor, DecisionTreeRegressor

```python
random_state = 12345
cv = 5
```

```python
model_linear = LinearRegression()

cross_val_score_rougher = cross_val_score(model_linear,  
                                              features_train, 
                                              target_rougher_train, 
                                              cv=cv, scoring=scorer).mean()
cross_val_score_final = cross_val_score(model_linear, 
                                            features_train, 
                                            target_final_train, 
                                            cv=cv, scoring=scorer).mean()

print('Оценка предсказаний модели линейной регрессии - эффективность обогащения чернового концентрата rougher.output.recovery:')
print(cross_val_score_rougher)
print('Оценка предсказаний модели линейной регрессии - эффективность обогащения финального концентрата final.output.recovery:')
print( cross_val_score_final)
```

```python
model_tree = DecisionTreeRegressor()

cross_val_score_rougher_tree = cross_val_score(model_tree, 
                                              features_train, 
                                              target_rougher_train, 
                                              cv=cv, scoring=scorer).mean()
cross_val_score_final_tree = cross_val_score(model_tree, 
                                            features_train, 
                                            target_final_train, 
                                            cv=cv, scoring=scorer).mean()

print('Оценка предсказаний модели решающего дерева - эффективность обогащения чернового концентрата rougher.output.recovery:')
print(cross_val_score_rougher_tree)
print('Оценка предсказаний модели решающего дерева - эффективность обогащения финального концентрата final.output.recovery:')
print( cross_val_score_final_tree)
```

Наблюдаем лучший результат у модели RandomForestRegressor. 
Подберем гиперпараметры, чтобы улучшить работу модели.

```python
%%time 
results = []

for depth in range(1,5):
    
    for estimator in range(10, 21, 5):
        
        model = RandomForestRegressor(random_state=random_state, 
                                       n_estimators=estimator, 
                                       max_depth=depth)
        
        scorer = make_scorer(smape, greater_is_better=False) 
    
        cross_val_score_rougher = cross_val_score(model, features_train, target_rougher_train, cv=cv, scoring=scorer).mean()
        cross_val_score_final = cross_val_score(model, features_train, target_final_train, cv=cv, scoring=scorer).mean()

        smape_final_result = final_smape(cross_val_score_rougher, cross_val_score_final)

        results.append({'model name': model.__class__.__name__, 
                        'Hyperparameters': {'random_state': random_state, 
                                            'n_estimators': estimator, 
                                            'max_depth':depth}, 
                        'smape_final_result': smape_final_result})

display(results)
```

Видим, что подбор гиперпараметров улучшил работу модели. Протестируем работу модели случайного леса с гиперпараметрами n_estimators = 10 и max_depth = 4 на тестовой выборке.
